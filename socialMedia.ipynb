```python
#increase in the users
import matplotlib.pyplot as plt
# Years
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024]

# Number of users
users = [970, 1000, 1400, 1800, 2100, 2500, 2900, 3400, 3900, 4400, 4900, 5300, 5800, 6200, 6700]

# Plot
plt.figure(figsize=(10, 6))
plt.plot(years, users, marker='o', linestyle='-')
plt.title('Increase in Social Media Users (2010-2024)')
plt.xlabel('Year')
plt.ylabel('Number of Users (in millions)')
plt.grid(True)
plt.xticks(years, rotation=45)
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('social.csv')
organization_mapping = {
    'University': 'University',
    'Private': 'Private',
    'School': 'School',
    'Company': 'Company',
    'Government': 'Government',
    'N/A': 'N/A'
}

data['5. What type of organizations are you affiliated with?'] = data['5. What type of organizations are you affiliated with?'].map(organization_mapping)
numerical_columns = ['10. How often do you get distracted by Social media when you are busy doing something?',
                     '12. On a scale of 1 to 5, how easily distracted are you?',
                     '13. On a scale of 1 to 5, how much are you bothered by worries?',
                     '15. On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?',
                     '17. How often do you look to seek validation from features of social media?',
                     '18. How often do you feel depressed or down?',
                     '19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?',
                     '20. On a scale of 1 to 5, how often do you face issues regarding sleep?']

organization_average = data.groupby('5. What type of organizations are you affiliated with?')[numerical_columns].mean()

# Plotting the bar chart
plt.figure(figsize=(12, 6))
organization_average.plot(kind='bar')
plt.title('Average Values by Organization Type')
plt.xlabel('Organization Type')
plt.ylabel('Average')
plt.xticks(rotation=0)
plt.legend(title='Problem')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Selecting the columns with numerical data
numerical_columns = ['10. How often do you get distracted by Social media when you are busy doing something?',
                     '12. On a scale of 1 to 5, how easily distracted are you?',
                     '13. On a scale of 1 to 5, how much are you bothered by worries?',
                     '15. On a scale of 1-5, how often do you compare yourself to other successful people through the use of social media?',
                     '17. How often do you look to seek validation from features of social media?',
                     '18. How often do you feel depressed or down?',
                     '19. On a scale of 1 to 5, how frequently does your interest in daily activities fluctuate?',
                     '20. On a scale of 1 to 5, how often do you face issues regarding sleep?']

# Calculating the correlation matrix
correlation_matrix = data[numerical_columns].corr()

# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Problems')
plt.show()
```
